#### Funkcje realizujące zadania:

```python
# Solution 1

# Empty functions to be implemented in cell below

def step_1a(word):
    pass

def step_1b(word):
    pass

def step_1b_extra(word):
    pass

def step_1c(word):
    pass

def step_2(word):
    pass

def step_3(word):
    pass

def step_4(word):
    pass

def step_5a(word):
    pass

def step_5b(word):
    pass
```

```python
# Implemented functions and extra functions that will help us in the implementations such as is_vowel,
# contains_vowel, count_m, check_v_condition, check_d_condition, check_o_condition, etc.

import re

vowels = 'aeiou'

def is_vowel(letter):
    return letter.lower() in vowels

def is_consonant(letter):
    return not is_vowel(letter)

def count_m(word):
    word = word.lower()
    m = 0
    in_vowel_sequence = False
    
    for i in range(len(word)):
        # Check for a vowel
        if is_vowel(word[i]):
            in_vowel_sequence = True
        # If now it's a consonant and before there was a vowel we have VC sequention, so m+1
        else:
            if(in_vowel_sequence):
                m += 1
                in_vowel_sequence = False
    
    return m

# Check if word contains a vowel (*v*), if it does then return True

def check_v_condition(word):
    word = word.lower()
    return any(is_vowel(letter) for letter in word)

# Check if word ends with a double consonant (*d), if it does then return True
    
def check_d_condition(word):
    word = word.lower()
    if((word[-1]==word[-2]) and is_consonant(word[-1])):
        return True
    else:
        return False
    
# Check if word ends with a consonant-vowel-consonant pattern (*o), where the second consonant is NOT X\W\Y, if it does then return True

def check_o_condition(word):
    word = word.lower()
    if (is_consonant(word[-3]) and is_vowel(word[-2]) and is_consonant(word[-1]) and word[-1] not in 'wxy'):
        return True
    else:
        return False
    
def regex_replace_suffix(word, rules):
    for pattern, replacement in rules:
        result = re.sub(rf'{pattern}$', replacement, word, flags=re.IGNORECASE)
        if pattern == replacement:
            return result
        if result != word:
            return result
    return word

def step_1a(word):
    rules=[
        ("sses","ss"),
        ("ies","i"),
        ("ss","ss"),
        ("s","")
    ]
    return regex_replace_suffix(word, rules)

def step_1b(word):
    # This step is easier to implement using if statements
    lowered_word=word.lower()
    if count_m(word)>0 and lowered_word.endswith('eed'):
        return word[:-1]
    if check_v_condition(word[:-2]) and lowered_word.endswith('ed'):
        return word[:-2]
    if check_v_condition(word[:-3]) and lowered_word.endswith('ing'):
        return word[:-3]
    
    # If none of the above, return word as it is
    return word
    
def step_1b_extra(word):
    # This step is also easier to implement using if statements
    lowered_word=word.lower()
    if lowered_word.endswith('at') or lowered_word.endswith('bl') or lowered_word.endswith('iz'):
        return word + 'e'
    if check_d_condition(word) and not lowered_word.endswith('l') and not lowered_word.endswith('s') and not lowered_word.endswith('z'):
        return word[:-1]
    if count_m(word)==1 and check_o_condition(word):
        return word + 'e'
    
    # If none of the above, return word as it is
    return word

def step_1c(word):
    # This step is also easier to implement using if statement
    if check_v_condition(word[:-1]) and word.lower().endswith('y'):
        return word[:-1]+'i'
    return word

def step_2(word):
    rules=[
        ("ational","ate"),
        ("tional","tion"),
        ("enci","ence"),
        ("anci","ance"),
        ("izer","ize"),
        ("abli","able"),
        ("alli","al"),
        ("entli","ent"),
        ("eli","e"),
        ("ousli","ous"),
        ("ization","ize"),
        ("ation","ate"),
        ("ator","ate"),
        ("alism","al"),
        ("iveness","ive"),
        ("fulness","ful"),
        ("ousness","ous"),
        ("aliti","al"),
        ("iviti","ive"),
        ("biliti","ble")
    ]
    # Every rule in step 2 requires m>0
    if count_m(word)>0:
        return regex_replace_suffix(word, rules)
    
    # If not, return word as it is
    return word

def step_3(word):
    rules=[
        ("icate","ic"),
        ("ative",""),
        ("alize","al"),
        ("iciti","ic"),
        ("ical","ic"),
        ("ful",""),
        ("ness","")
    ]
    if count_m(word)>0:
        return regex_replace_suffix(word, rules)
    return word
    
def step_4(word):
    # I didn't include to the rules rule for ION because it needs extra condition for stem so
    # I will implement it as separate if statement
    rules=[
        ("al",""),
        ("ance",""),
        ("ence",""),
        ("er",""),
        ("ic",""),
        ("able",""),
        ("ible",""),
        ("ant",""),
        ("ement",""),
        ("ment",""),
        ("ent",""),
        ("ou",""),
        ("ism",""),
        ("ate",""),
        ("iti",""),
        ("ous",""),
        ("ive",""),
        ("ize","")
    ]
    # Every rule in step 4 requires m>1
    if count_m(word)>1:
        if word.lower().endswith('ion'):
            if word[:-3].lower().endswith('t') or word[:-3].lower().endswith('s'):
                return word[:-3]
        else:
            result = regex_replace_suffix(word, rules)
            if result != word:
                return result
    
    # If not, return word as it is
    return word

def step_5a(word):
    # This step is also easier to implement using if statements
    if count_m(word)>1 and word.lower().endswith('e'):
        return word[:-1]
    if count_m(word)==1 and not check_o_condition(word) and word.lower().endswith('e'):
        return word[:-1]
    return word

def step_5b(word):
    # This step is also easier to implement using if statements
    if count_m(word)==1 and check_d_condition(word) and word.lower().endswith('l'):
        return word[:-1]
    return word
```

```python
def stem_word(word):
    # Process the word through all the steps
    steps=[step_1a, step_1b, step_1b_extra, step_1c, step_2, step_3, step_4, step_5a, step_5b]
    for step in steps:
        word=step(word)
        print(step, word)
    return word
```
